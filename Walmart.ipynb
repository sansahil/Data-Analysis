### Q**2: What is the average unit price of all products sold ?**

 Assists in price optimization for different product lines to maximize profit margins.

```python
average_price = df.groupby("product_name")['unit_price'].mean()
average_price
```

```python
 # calculate average price
average_price = df.groupby("product_name")['unit_price'].mean()

# plot bar graph

plt.figure(figsize=(10,6))
bar = average_price.plot(kind ='bar', color ='skyblue')
plt.title('Average price per Product')
plt.ylabel('Average Name')
plt.xlabel('Average unit price')
plt.xticks(rotation = 45, ha='center')

```

```python
#calculate average price
average_price = df.groupby("product_name")['unit_price'].mean()

# plot bar graph

plt.figure(figsize=(10,6))
bar = average_price.plot(kind ='bar', color ='skyblue')
plt.title('Average price per Product')
plt.ylabel('Average Name')
plt.xlabel('Average unit price')
plt.xticks(rotation = 45, ha='center')


# Add data lable

for index, value in enumerate(average_price):
  plt.text(index, value, f'{value:2f}', ha='center' , va = 'bottom', fontsize= 10)


plt.tight_layout()
plt.show()
```

### Q4: **Which stores are underperforming in terms of sales volume?**

Allow the operation team to allocate more strategically and support underperforming location

```python
sales_by_store = df.groupby('store_id')['quantity_sold'].sum().sort_values()
sales_by_store
```

```python
sales_by_store.median()
```

```python
# Define underperforming stores as below a threshold (e.g , below the median)

threshold = sales_by_store.median()
underperforming_stores = sales_by_store[sales_by_store < threshold]

# Display underperforming stores

print("underperforming stores:")
print(underperforming_stores)
```

### **Q6: Which product categoies see the largest increase in demand during holidays ?**

Enables better inventory planning for peak holiday seasons.

```python
# filter data for holidays
holiday_sales = df[df['holiday_indicator']== True]

# Group by product_name and calculate total quantity sold
holiday_demand = holiday_sales.groupby('product_name')['quantity_sold'].sum()


# Display result
print(holiday_demand)


# plot thr bar graph

plt.figure(figsize=(12,6))
ax = holiday_demand.plot(kind = 'bar', color= 'skyblue' , edgecolor ='black')
```
